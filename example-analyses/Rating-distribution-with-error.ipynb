```python
import pandas as pd
```

```python
contest = '586'
# contest = '587'
df = pd.read_csv(f'../contests/summaries/{contest}_summary_KLUCB.csv')
df['rank'] = df.index + 1
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df.plot(x='rank', y='score', yerr='precision', style='_')
plt.xlim(0, 20)
plt.ylim(1.75, 2.15)
plt.title(f'Cartoon caption contest {contest}')
plt.ylabel('Average score')
plt.xlabel('Rank')
```
